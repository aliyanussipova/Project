# Visualizing Earnings Based on College Majors

## Introduction

In this project, we will explore the dataset of the job outcomes of students who graduated from college between 2010 and 2012 using vizualisations. The original data was released by American Community Survey and cleaned by FiveThirtyEight [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors)).

Each row of the dataset represents a different college major and contains a specific information on the gender, unemployment rate, median salareis, type of employement, and etc.

* Rank - Rank by median earnings (the dataset is ordered by this column). 
* Major_code - Major code. 
* Major - Major description.
* Major_category - Category of major.
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time.
* Men - Male graduates.
* Women - Female graduates.
* ShareWomen - Women as share of total.
* Employed - Number employed.
* Median - Median salary of full-time, year-round workers.
* Low_wage_jobs - Number in low-wage service jobs.
* Full_time - Number employed 35 hours or more.
* Part_time - Number employed less than 35 hours.

Throughout the project we are aimed to find out which majors have higher salaries, which majors are predominantly male or female, and which major categories have most of students.

## Scatter Plots

Now, let's start exploring the dataset using the visualizations. We will start with scatter plots first to explore whether students in more poplular majors make more money, if majors with predominantly female students make more money or if there is a correlation between the full-time employment and the salary amount.
The analysis will be made based on the follwing relations:
* Sample_size and Median
* Sample_size and Unemployment_rate
* Full_time and Median
* ShareWomen and Unemployment_rate
* Men and Median
* Women and Median

```python
#Scatter Plot 1
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Median vs. Sample_size')
```

```python
#Scatter Plot 2
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment_rate vs. Sample_size')
```

```python
#Scatter Plot 3
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full_time vs. Median')
```

```python
#Scatter Plot 4
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('ShareWomen vs. Unemployment_rate')
```

```python
#Scatter Plot 5
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')
```

```python
#Scatter Plot 6
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
```

**Observations.**
A high-level observation presents that all scatter plots have a weak negative correlation with similar trend, except for scatter plot 4 "ShareWomen vs. Unemployment_rate". Therefore, no meaningful conclusions can be made on the ShareWomen vs. Unemployment_rate relation.
Scatter plot 1 on the relationship between the median salary and our sample size shows that the majority of the salary is ranging between 25,000 - 40,000 USD for the sample size being below 1000 people.
Scatter plot 2 provides information that the unemployment rate within our sample size mostly falls between 0.05 to 0.1.
Scatter plot 3 also shows the weak correlation between working full-time and their median salaries. 
Comparing scatter plots 5 and 6 on Men, Women vs.Median, we can observe the similar trend on both plots, with the vast mjority falling between 20,000 USD and 45,000 USD.

## Histograms

We will explore the distributions of values in columns using histograms. 

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen"]
fig = plt.figure(figsize=(7,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist')
    ax.set_title(cols[r])
```

Based on the data visualization above, we can make a conclusion that the most common median salary is in a range between 30,000 USD - 40,000 USD.
Also, there are at least 28 majors (16.27% of total in the given dataset) that have predominatly female students (70% women share)and 24 majors (13.95% of total) with 60% of female students. 

```python
cols = ["ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(7,12))
for r in range(1,4):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist')
    ax.set_title(cols[r])
```

The most common unemployment rate based on the fisrt histogram above is between 0.06 and 0.07. Moreover, there are overall more female graduates compared to male graduates of the given dataset.

## Scatter Matrix Plot

In order to explore potential relationships and distributions simultaneously we will create a scatter matrix plot that combines both scatter plots and histograms into one grid of plots. 

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Women', 'Men']], figsize=(10,10))
```

Based on the scatter matrix relationship between Men and Women, we can observe a weak positive correlation. A similar weak correlation could be observed between Sample_size and the Median scatter matrix below.

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

The 3X3 matrix above does show that we do not have any distinctive linear correlations between our variables. Although, there is ceratainly better correlation between Sample_size & Median, and Sample_size & Unemployment_rate. Other correlations are too weak.

## Bar Plots

Now, let's explore data using bar plots to compare the percentages of women (ShareWomen) from the first ten rows and last ten rows of the recent_grads dataframe, as well as  the unemployment rate (Unemployment_rate) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
recent_grads.head(10).tail(10).plot.bar(x='Major', y='ShareWomen')
```

The chart above shows that majors such as Astronomy and Astrophysics, and Actuarial Science have the highest women share among others, approximately 45% and 52% respectively. While, the predominatly women major will be only Astronomy and Astrophysics. Other majors are predominantly male, including all the engineering majors. However, this chartb represents only top and bottom 10 rows, therefore, we cannot relate it to the whole population.
But, knowing, that the rows re ranked based on the medium salaries, we can make a conclusion that men graduates are earning more on average (as the first top majors in the dataset are engineering majors). 

```python
recent_grads.head(10).tail(10).plot.bar(x='Major', y='Unemployment_rate', legend =False)
```

The highest unemployment rate from our selected data is Nuclear Engineering major with only 0.18. The overall level of the unemployment rate is quite low, and, surprisingly, majors with the least medium salary are having lower unemployment compared to engineering majors.

## Grouped Bar Plot

In order to compare the number of men with the number of women in each category of majors we will use the grouped bar plot.

```python
recent_grads.groupby('Major_category')['Men', 'Women'].sum().plot(kind='bar')
```

The highest number of men and women are presented in Business majors, while the least amount is in Interdisciplinary majors. The predominantly female majors are Education, Health, abd Psychology & Social Work majors. Predominantly male majors are Computer & Mathematics and Engineering.
